```python
import pandas as pd

# Load the data from the Excel file
file_path = 'client_returns_fact.xlsx'
df = pd.read_excel(file_path)

# Print the column names to debug
print("Column names:", df.columns)

# Mapping from client_name to (Strategy, Benchmark)
mapping = {
    "Warren Miller": ("Equity", "S&P 500"),
    "Sandor Clegane": ("Government Bonds", "Bloomberg Barclays US Aggregate Bond Index"),
    "Hari Seldon": ("High Yield Bonds", "ICE BofAML US High Yield Index"),
    "James Holden": ("Leveraged Loans", "S&P/LSTA Leveraged Loan Index"),
    "Alice Johnson": ("Commodities", "Bloomberg Commodity Index"),
    "Bob Smith": ("Private Equity", "Cambridge Associates Private Equity Index"),
    "Carol White": ("Long Short Equity Hedge Fund", "HFRI Equity Hedge Index"),
    "David Brown": ("Long Short High Yield Bond", "HFRI Fixed Income - Credit Index")
}

# Function to map client_name to Strategy and Benchmark
def map_strategy_benchmark(client_name):
    return mapping.get(client_name, ("Unknown", "Unknown"))

# Ensure to use the correct column name for client names
if 'client_name' not in df.columns:
    raise KeyError("The DataFrame does not contain a 'client_name' column.")

# Apply the mapping function
df[['Strategy', 'Benchmark']] = df['client_name'].apply(lambda x: pd.Series(map_strategy_benchmark(x)))

# Convert the DataFrame to a dictionary
data = df.to_dict(orient='list')

# Save the dictionary to a Python file
with open('data_dict.py', 'w') as file:
    file.write(f"data = {data}")

print("Data has been saved to data_dict.py")

```
